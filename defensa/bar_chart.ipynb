```python
# i need to plot a histogram with 5 bars where there are two bars with lower values and three bars with higher values

import matplotlib.pyplot as plt
# dpi
plt.figure(dpi=300)


data = [400, 430, 300, 240, 440]
# x label 1, 2, 3, 4, 5
plt.xticks(range(len(data)), range(1, len(data)+1))
plt.bar(range(len(data)), data)
#xlabel
plt.xlabel('Naturalidad')
plt.ylabel('Frecuencia')

plt.show()

```

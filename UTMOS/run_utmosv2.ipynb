```python
import pandas as pd
import utmosv2

test_set = pd.read_csv('test_shuffled.csv')
model = utmosv2.create_model(pretrained=True)

#iterate over the test set, for every row, run inference on the path indicated in the 'stimuli' column, and save to new column 'mos_pred

test_set['mos_pred'] = test_set['stimuli'].apply(lambda x: model.predict(input_path=x))

#save the test set with the new column to a new csv file

test_set.to_csv('test_shuffled_with_predictions.csv', index=False)
```

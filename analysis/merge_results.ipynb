```python
import pandas as pd
```

# Open and drop empty answers

```python
# open results csv file

results = pd.read_csv('results_processed.csv')
results.head()

# # if rating column is empty, delete row
results = results.dropna(subset=['score'])

# save results to csv
results.to_csv('results_processed.csv', index=False)
```

### Combine participants and results_processed dataframes

```python
# Combine participants and results_processed dataframes

participants = pd.read_csv('participants.csv')

# merge dataframes
merged = pd.merge(participants, results, on='participant_id')

# save merged dataframe to csv
merged.to_csv('merged.csv', index=False)
```

### Combine merged dataframes with the the tts atributes dataframe

```python
merged = pd.read_csv('merged.csv')
merged.head()
```

```python
# add stimuli group to merged dataframe

stimuli_group = []

# read the stimuli code in the merged dataframe
for item in merged['stimuli']:
    stimuli_group.append(item.split('/')[1])

# add stimuli group to merged dataframe
merged['stimuli_group'] = stimuli_group

# save merged dataframe to csv

merged.to_csv('merged.csv', index=False)
```

```python
tts_atributes = pd.read_csv('tts_atributes.csv')

# merge dataframes

merged_final = pd.merge(merged, tts_atributes, on='stimuli_group')

# save merged_final dataframe to csv

merged_final.to_csv('merged_final.csv', index=False)
merged_final.head()
```
